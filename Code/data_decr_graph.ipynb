```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from datetime import datetime
```

```python
df_price = pd.read_csv('full_close 1_date.csv')
df_price['Date'] = pd.to_datetime(df_price['Date'])
df_price.set_index(['Date'], inplace = True)
df_price.head()
df_price.shape
```

```python
#Daily Prices

fig, ax = plt.subplots(1, figsize=(15,10))
dates = df_price.index
ax.plot(dates, df_price["PFE_Close"], color = 'yellow', label = "PFE")
ax.plot(dates, df_price["JNJ_Close"], color = 'blue', label = "JNJ")
ax.plot(dates, df_price["MRK_Close"], color = 'orange', label = "MRK")
ax.plot(dates, df_price["NVS_Close"], color = 'black', label = "NVS")

ax.set_ylabel('Daily Prices')
ax.set_xlabel('Time')
ax.set_title('Daily Prices')

plt.legend(loc="upper left")
sns.despine(ax=ax)

plt.savefig('Daily Prices.jpg')
plt.show()
```

```python
#Net Return

df_return = pd.read_csv('net return_date.csv')
df_return['Date'] = pd.to_datetime(df_return['Date'])
df_return.set_index(['Date'], inplace = True)
```

```python

fig, ax = plt.subplots(1, figsize=(30,10))
dates = df_return.index
ax.plot(dates, df_return["PFE_Close"], color = 'yellow', label = "PFE")
ax.plot(dates, df_return["JNJ_Close"], color = 'blue', label = "JNJ")
ax.plot(dates, df_return["MRK_Close"], color = 'orange', label = "MRK")
ax.plot(dates, df_return["NVS_Close"], color = 'black', label = "NVS")
ax.plot(dates, df_return["SP500_Close"], color = 'grey', label = "SP 500")

ax.set_ylabel('Daily Net Returns')
ax.set_xlabel('Time')
ax.set_title('Daily Net Returns')

plt.legend(loc="upper left")
sns.despine(ax=ax)

plt.savefig('Daily Net Returns.jpg')
plt.show()
```
